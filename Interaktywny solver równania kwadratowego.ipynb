### Interaktywny solver równania kwadratowego


```python
global_a :float = 0
global_b :float = 0
global_c :int = 0
```

```python
def show():
    # Path: Interaktywny wykres funkcji kwadratowej.ipynb
    import matplotlib.pyplot as plt
    import numpy as np

    plt.figure(figsize=(10,8), dpi=100)

    x=np.arange(-10,10,0.1)
    y=global_a*(x**2) + global_b*x + global_c

    plt.xlabel('oś x')
    plt.ylabel('oś y')
    plt.axhline(y=0, color="#cccccc")
    plt.axvline(x=0, color="#cccccc")
    plt.title(f'wykres y={global_a}*x^2 + {global_b}*x + {global_c}')

    plt.plot(x,y)
    plt.show()
```

```python
def solve(a: float = 1, b: float = 1, c :int = 0):
    global global_a 
    global_a = a
    global global_b
    global_b = b
    global global_c
    global_c = c
    """Rozwiązuje równanie kwadratowe ax^2 + bx + c = 0"""
    delta = b**2 - 4*a*c
    show()
    try:
        if delta < 0:
            return "Brak rozwiązań"
   
        elif delta == 0:
            return -b/(2*a)
        else:
            return (-b - delta**0.5)/(2*a), (-b + delta**0.5)/(2*a)
    except ZeroDivisionError:
        return "Brak rozwiązań"
    
    
```

### Prezentacja działania

```python
# Next show that it works
solve(a=1,b=5,c=6)
```

```python
solve(a=1,b=5)
```
