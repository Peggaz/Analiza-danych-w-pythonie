<h2>Import bibliotek i wczytanie danych</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

<h2>Pobranie i wyświetlenie informacji o liczbie kolumn i wierszy:</h2>

```python
print("Liczba kolumn",dt.shape[1])
print("Liczba wierszy",dt.shape[0])
dt = pd.read_csv("dane.csv")
print(dt)
```

<h2>Wartość średnia</h2> 

```python
print(dt.mean())
```

<h2>Odchylenie standardowe 

```python
print(dt.std())
```

<h2>Mediana

```python
print(dt.median())
```

<h2>Wartość maksymalna

```python
print(dt.max())
```

<h2>Wartość minimalna

```python
print(dt.min())
```

<h2>Macierz korelacji dla zbioru danych.

```python
correlation_matrix = dt.corr().round(2)
print(correlation_matrix)
```

<h2>Wykres korelacji wartości dla każdej z par kolumn.

```python
sns.heatmap(correlation_matrix, annot=True, cmap="RdYlBu")
plt.title("Macierz korelacji")
plt.show()
```

<h2>Histogram rozkładu wartościdla wszystkich kolumn

```python
for column in dt.columns:
    plt.hist(dt[column])
    plt.xlabel(column)
    plt.ylabel("Liczba wystąpień")
    plt.title(f"Histogram rozkładu wartości - {column}")
    plt.show()
```
