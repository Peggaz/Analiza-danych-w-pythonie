```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (50, 15)

print("Projekt opartyo dane ilości zgonów na COVID19 z danych prezentowanych na zajęciach")
```

```python
dane = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/archived_data/archived_time_series/time_series_19-covid-Deaths_archived_0325.csv')
dane = dane.groupby('Country/Region', as_index=False).sum()
#usuniecie niepotrzebnych kolumn
dane = dane.drop(['Lat', 'Long'], axis=1)
print("Pełna tabela z pogrupowanymi danymi wg krajów")
dane

```

```python

ostatni_dzien = dane[['Country/Region', dane.columns[-1]]]
posortowane_po_ilosci_zgonow = ostatni_dzien.sort_values(ostatni_dzien.columns[-1], ascending = False)
print(f'Dane dla {len(posortowane_po_ilosci_zgonow)} krajów posortowane od najmniejszej liczby zgonów')
posortowane_po_ilosci_zgonow
```

```python
usrednione_dane = dane.mean()
usrednione_dane_t = usrednione_dane.T
usrednione_dane_t.plot(xlabel='Data', ylabel='Ilość zgonów', title='średnia ilość zgonówper kraj')
plt.show()
print(usrednione_dane_t)
```

```python
najwiecej_zgonow = dane.loc[dane['Country/Region'].isin(posortowane_po_ilosci_zgonow.head(5)['Country/Region'].values)]
najwiecej_zgonow = najwiecej_zgonow.set_index('Country/Region')
najwiecej_zgonow_t = najwiecej_zgonow.T
najwiecej_zgonow_t.plot(xlabel='Data (M/D/Y)', ylabel='Ilość zgonów', title='Ilość wszystkich zgonów dla poszczególnych krajów')
plt.show()
najwiecej_zgonow.head(5)
```

```python

usrednione_dane_zgonow_dziennie = usrednione_dane_t
#tutaj sie cosrozjezdza jak wywolamy kilka razy pod rząd, dla bezpieczenstwa lepiej zawsze wywolywac caly oplik na raz a nie pojedyncze fragmenty
for it in range(len(usrednione_dane_t)):
    if it > 0:
        usrednione_dane_zgonow_dziennie[it] = usrednione_dane_t[it] - usrednione_dane_t[it-1]
    else: 
        usrednione_dane_zgonow_dziennie[it] = usrednione_dane_t[it]

usrednione_dane_zgonow_dziennie.plot(xlabel='Data (M/D/Y)', ylabel='Ilość zgonów', title='zachorowań w danym dniu na COViD-19')
plt.show()
usrednione_dane_zgonow_dziennie




    
        
```
